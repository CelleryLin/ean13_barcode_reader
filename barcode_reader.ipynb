```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import image_proc as ip

%load_ext autoreload
%autoreload 2
```

```python
img = cv2.imread("./test1.bmp")
img = cv2.resize(img, (300, 400))
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = ip.normalize(img)

# contrast enhancement
# img = ip.contrast(img,0)

# detect barcode
_, label, location = ip.barcode_detection(img, dilate_size=9, erode_size=7, morph_iter=4)
# binarization
# img = ip.binarize(img, 0.2)

plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")
plt.subplot(1,2,2)
plt.imshow(label*img, cmap="gray")
print(location)
plt.plot(location)
plt.show()
```

```python
# label

# find area
area = 0
bar_x = 0
bar_y = 0
for i in range(label.shape[0]):
    for j in range(label.shape[1]):
        if label[i,j] == 1:
            area += 1

for i in range(label.shape[0]):
    for j in range(label.shape[1]):
        if label[i,j] == 1:
                bar_x += i


for i in range(label.shape[0]):
    for j in range(label.shape[1]):
        if label[i,j] ==1.0:
            bar_y += j

bar_x/=area
bar_y/=area


print(area)
print(bar_x, bar_y)
```
